```python
data = pd.read_csv('./seoul_general_edu_list.csv')
data = data.iloc[:, :8]
```

```python
subset = data[data['대학교'] == '한양대학교'].dropna(subset='수강인원')
subset['수강인원'] = [int(n.split('/')[0]) for n in subset['수강인원']]
data.iloc[subset.index] = subset
```

```python
subset = data[data['대학교'] == '숭실대학교'].dropna(subset='교양영역')
subset['교양영역'] = [s.split('\n') for s in subset['교양영역']]
```

```python
for s in subset['교양영역']:
    if len(s) == 1:
        continue
    else:
        print(s)
        break
```

```python
for i, r in subset.iterrows():
    d = r['교양영역'].split('\n')
    if (len(d) == 1):
        continue
    else:
        temp = [s.split(']') for s in d]
        for t in temp:
            if t[0] == '[‘23이후':
                r['교양영역'] = t[1]
                print(r)
```

```python
data.to_csv('./total.csv', encoding='utf-8', index=False)
```
